Imports

```python
import selenium.webdriver as webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv
import time
import pandas as pd
import pickle
```

Stock Ticker Symbol Scraping

```python
with open(r'C:\Users\marti\Desktop\School\Year 4\Semester 4.1 - ISEP\30590 - ADVANCED PYTHON PROGRAMMING FOR ECONOMICS, MANAGEMENT AND FINANCE\Project\scraped_tickers.csv', 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerow(['SYMBOL', 'COMPANY NAME'])
    url = 'https://stockanalysis.com/stocks/'
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    driver = webdriver.Chrome(options=chrome_options)
    driver.get(url)
    cancel_button = WebDriverWait(driver, 30).until(EC.element_to_be_clickable((By.XPATH, "/html/body/div/div[2]/div/button")))
    driver.execute_script("arguments[0].click();", cancel_button)
    for n in range(12):
        symbols = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="main-table"]/tbody//td[1]/a')))
        company_names = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="main-table"]/tbody//td[2]')))
        for symbol, company_name in zip(symbols, company_names):
            csvwriter.writerow([symbol.text, company_name.text])
        next_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="main"]/div/div/nav/button[2]')))
        driver.execute_script("arguments[0].click();", next_button)
        time.sleep(0.5)
    driver.quit()

print('Tickers successfully exported to CSV!')
```

Create & Save a Name to Ticker Dictionary

```python
df = pd.read_csv("scraped_tickers.csv", encoding='cp1252')
df = df.astype(str)
print(len(df))
```

```python
df.head()
```

```python
name2ticker_dict = {name.upper(): symbol.upper() for name, symbol in zip(df['COMPANY NAME'], df['SYMBOL'])}
print(name2ticker_dict)
print(len(name2ticker_dict)) # there are 48 companies with duplicate ticker symbols (49 extra duplicates)
```

```python
with open('name2ticker_dict.pkl', 'wb') as f:
    pickle.dump(name2ticker_dict, f)
```
